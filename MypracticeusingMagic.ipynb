```python
#if there waws header we could do this
#magic_data = pd.read_csv('magic04.data', index_col= 0)
cols = ["fLength", "fWidth", "fSize", "fConc", "fConc1", "fAysm", "fM3Long", "fM3Trans", "fAlpha", "fDist", "class"]
magic_data = pd.read_csv('magic04.data', names = cols)
magic_data.head()
```

shape of dataset

```python
magic_data.shape
```

```python
magic_data["class"] = magic_data["class"].replace({"g": 1, "h": 0})
```

```python
class_counts = magic_data['class'].value_counts()
print(class_counts)
```

```python
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
```

```python
train, valid, test = np.split(magic_data.sample(frac=1), [int(0.8*len(magic_data)), int(0.9*len(magic_data))])
```

```python
def scale_dataset(dataframe, oversample=False, undersample=False):
  X = dataframe[dataframe.columns[:-1]].values
  y = dataframe[dataframe.columns[-1]].values

  scaler = StandardScaler()
  X = scaler.fit_transform(X)

  if oversample:
    ros = RandomOverSampler()
    X, y = ros.fit_resample(X, y)
  if undersample:
    rus = RandomUnderSampler()
    X, y = rus.fit_resample(X, y)

  data = np.hstack((X, np.reshape(y, (-1, 1))))

  return data, X, y
```

```python
Train, x_train, y_train = scale_dataset(train, oversample=True, undersample=True)
Valid, x_valid, y_valid = scale_dataset(valid, oversample=False,  undersample=False)
Test, x_test, y_test = scale_dataset(test, oversample=False,  undersample=False)
```

KNN

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
##no of neighbours is making the results better
knn_model.fit(x_train, y_train)
```

```python
knn_model.score(x_train, y_train)
```

```python
y_pred = knn_model.predict(x_valid)
```

```python
print(classification_report(y_valid, y_pred))
```

```python
y_pred_test = knn_model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_test))
```
